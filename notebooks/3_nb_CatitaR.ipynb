```python
df_sii = pd.merge(df2019, df2020, how='outer')
```

```python
df_sii = df_sii.melt(id_vars=['year','Día'], value_vars=['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic','Día'], var_name=['mes'], value_name='valor' )
```

```python
print(type(df_sii))
df_sii.columns
```

```python
temp_column = df_sii.pop(('Día'))
```

```python
df_sii.insert( 2, 'dia', temp_column)
```
